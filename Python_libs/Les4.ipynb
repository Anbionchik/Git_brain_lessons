### Задание №2

```python
t = np.linspace(0,10,51)
f = np.cos(t)
```

```python
plt.plot(t,f, color='green')

plt.title(
    "График f(t)",
    fontsize=16,
    fontweight="bold",
    color="black",
    family="calibri",
)
label_font = {
    "fontsize": 13,
    "family": "calibri",
}
plt.xlabel('Значения t', fontdict=label_font)
plt.ylabel('Значения f', fontdict=label_font)
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

### Задание №3

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace = 0.3, hspace = 0.3)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_xlim(-5, 5)
ax2.plot(x, y2)
ax1.set_title("График y2")
ax3.plot(x, y3)
ax1.set_title("График y3")
ax4.plot(x, y4)
ax1.set_title("График y4")
```

### Задание №4

```python
fraud_df = pd.read_csv('creditcard.csv')
fraud_df.tail()
```

```python
class_counts = fraud_df.value_counts(subset='Class')
class_counts.plot(kind='bar')
```

```python
class_counts.plot(kind='bar', logy=True)
```

```python
fraud_df[fraud_df["Class"] == 0]["V1"].hist(bins = 20, density = True, alpha = 0.5)
```

```python
fraud_df[fraud_df["Class"] == 0]["V1"].hist(bins = 20, density = True, alpha = 0.5, color='gray')
fraud_df[fraud_df["Class"] == 1]["V1"].hist(bins = 20, density = True, alpha = 0.5, color='red')
```

```python
plt.hist(fraud_df[fraud_df["Class"] == 0]["V1"], bins = 20, density = True, alpha = 0.5, color='gray')
plt.hist(fraud_df[fraud_df["Class"] == 1]["V1"], bins = 20, density = True, alpha = 0.5, color='red')
plt.legend(labels=["Class 0", "Class 1"])
plt.ylabel("V1", font='calibri')
plt.show()
```
