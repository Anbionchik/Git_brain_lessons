```python
def circle(x0, y0, R):
    x = np.linspace((x0-R), (x0+R), 101)
    y1 = np.sqrt((R)**2 - (x-x0)**2) + y0
    y2 = -(np.sqrt((R)**2 - (x-x0)**2)) + y0
    plt.figure(figsize=(5,5))
    plt.plot(x,y1)
    plt.plot(x,y2)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
circle(5,5,10)
```

```python
def elips(x0, y0, R1, R2):
    x = np.linspace((x0-R2), (x0+R2), 101)
    y1 = np.sqrt(((R1**2) * ((R2**2) - (x - x0)**2))/R2**2) + y0
    y2 = -(np.sqrt(((R1**2) * ((R2**2) - (x - x0)**2))/R2**2)) + y0
    plt.figure(figsize=(5,5))
    if R1>R2:
        plt.axis([x0-R1,x0+R1,y0-R1,y0+R1])
    else:
        plt.axis([x0-R2,x0+R2,y0-R2,y0+R2])
    plt.plot(x,y1)
    plt.plot(x,y2)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
elips(1000,100,300,30)
```

```python
def circle_polar(R):
    x = np.linspace(0, np.pi*2, 101)
    y = np.empty(101)
    y.fill(R)
    plt.polar(x,y)
    plt.show()
```

```python
circle_polar(5.5)
```

```python
def line_polar(x1,y1,x2,y2):
    x = np.linspace(x1, x2, 101)
    y = np.linspace(y1, y2, 101)
    plt.polar(x,y)
    plt.show()
```

```python
line_polar(5,4,-1,8)
```

#### Задание №4

```python
x = np.linspace(-2,5,500)
y1 = x**2 - 1
y2 = 1-(1-np.exp(x))/x
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
def equations(p):
    x,y = p
    return (x**2 - 1 - y, 1-(1-np.exp(x))/x - y)
points = [(-1,2),(2.5,5),(4.5,15)]
answers=[]
for point in points:
    answers.append(fsolve(equations, point))

for n, answ in enumerate(answers):
    print(f'Корень №{n+1}: ({answ[0]},{answ[1]})')
```

```python
x = np.linspace(-2,5,500)
y1 = np.exp(x) + 2*x - x**3 - 1
y2 = x*0
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
def equations(p):
    x,y = p
    return (np.exp(x) + 2*x - x**3 - 1 - y, 0 - y)
points = [(-1.5,0),(0,0),(2.5,0),(4,0)]
answers=[]
for point in points:
    answers.append(fsolve(equations, point))

for n, answ in enumerate(answers):
    print(f'Корень №{n+1}: ({answ[0]},{answ[1]})')
```
